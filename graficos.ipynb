```python
import pandas as pd
import matplotlib.pyplot as plt

proyectos = pd.read_csv('proyectos.csv')
```

```python
datos = pd.DataFrame({
    "pg": [795, 1199, 1402.6, 2128, 3500, 4350],
    "pp": [None, 10, 15, 20, 30, 45],
    "tpv": [None, 86, 165, 277, 456, 348],
    "tpf": [None, 65, 104, 153, 151, 148],
    "tpe": [None, 39, 65, 69, 62, 81],
    "pa": [1178, 6691, 10108, 18813, 26537, None],
    "po": [220286, 220803, 222376, 225756, 212296, None]
})
%matplotlib inline
```

```python
datos['pg_p'] = datos['pg']
datos['pp_p'] = datos['pp']

# Calcular evolución porcentual
for i in range(1, len(datos)):
    datos.loc[i, 'pg_p'] = datos.loc[i, 'pg'] * 100 / datos.loc[i-1, 'pg']
    datos.loc[i, 'pp_p'] = datos.loc[i, 'pp'] * 100 / datos.loc[i-1, 'pp']
```
